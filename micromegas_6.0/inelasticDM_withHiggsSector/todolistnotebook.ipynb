```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('doc/info1.csv')
df2 = pd.read_csv('doc/info2.csv')
df3 = pd.read_csv('doc/info3.csv')
df4 = pd.read_csv('doc/info4.csv')
df5 = pd.read_csv('doc/info5.csv')
df6 = pd.read_csv('doc/info6.csv')
```

```python
def conversion(df): 
    # Inicializar listas para almacenar los valores de cada arreglo
    arreglo1 = []
    arreglo2 = []
    arreglo3 = []
    arreglo4 = []

    # Iterar a través del DataFrame
    for indice, fila in df.iterrows():
        # Organizar los valores en los arreglos según el patrón especificado
        if indice % 4 == 0:
            arreglo1.append(fila)
        elif indice % 4 == 1:
            arreglo2.append(fila)
        elif indice % 4 == 2:
            arreglo3.append(fila)
        elif indice % 4 == 3:
            arreglo4.append(fila)

    # Crear nuevos DataFrames con las filas agrupadas
    df_arreglo1 = pd.DataFrame(arreglo1)
    df_arreglo2 = pd.DataFrame(arreglo2)
    df_arreglo3 = pd.DataFrame(arreglo3)
    df_arreglo4 = pd.DataFrame(arreglo4)
    return df_arreglo1, df_arreglo2, df_arreglo3, df_arreglo4
```

```python
d1,d2,d3,d4 = conversion(df1)
e1,e2,e3,e4 = conversion(df2)
f1,f2,f3,f4 = conversion(df3)
g1,g2,g3,g4 = conversion(df4)
h1,h2,h3,h4 = conversion(df5)
i1,i2,i3,i4 = conversion(df6)
```

```python
plt.figure(figsize=(12,7.5))
plt.plot(d1['Mchi1'], d1['Densidad_reliquia'],lw=2,label='Mh = 0.1 , MA = 0.1Mchi')
plt.plot(d2['Mchi1'], d2['Densidad_reliquia'],lw=2,label='Mh = 0.1 , MA = 1Mchi')
plt.plot(d3['Mchi1'], d3['Densidad_reliquia'],lw=2,label='Mh = 0.1 , MA = 2Mchi')
plt.plot(d4['Mchi1'], d4['Densidad_reliquia'],lw=2,label='Mh = 0.1 , MA = 4Mchi')
plt.axhline(y=0.12, color='blue', linestyle='--', label='Valor teorico de la densidad reliquia')
plt.xlabel('$M_{\chi} GeV$',size=30)
plt.ylabel('$\Omega h$',size=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel(r'$M_{\chi}\rm(GeV)$',size=30)
plt.ylabel(r'$\Omega_{s}$',size=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.legend(fontsize=15)
plt.show()
```
